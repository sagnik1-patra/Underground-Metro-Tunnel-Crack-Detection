```python
import os
import cv2
import json
import yaml
import pickle
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from pdf2image import convert_from_path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint


# ===========================================================
# FIXED PATHS
# ===========================================================
pdf_path = r"C:\Users\NXTWAVE\Downloads\Underground Metro Tunnel Crack Detection\7261049\16. Tzika et al..pdf"
output_dir = r"C:\Users\NXTWAVE\Downloads\Underground Metro Tunnel Crack Detection"
images_dir = os.path.join(output_dir, "extracted_frames")
os.makedirs(images_dir, exist_ok=True)

POPPLER_PATH = r"C:\Users\NXTWAVE\Downloads\Release-25.11.0-0\poppler-25.11.0\Library\bin"


# ===========================================================
# 1. EXTRACT IMAGES FROM PDF
# ===========================================================
print("[INFO] Extracting images from PDF...")

pages = convert_from_path(pdf_path, dpi=300, poppler_path=POPPLER_PATH)

image_paths = []
for idx, page in enumerate(pages):
    img_path = os.path.join(images_dir, f"page_{idx}.jpg")
    page.save(img_path, "JPEG")
    image_paths.append(img_path)

print(f"[INFO] Extracted {len(image_paths)} images.")


# ===========================================================
# 2. PREPROCESS IMAGES
# ===========================================================
X = []
y = []

print("[INFO] Preprocessing images...")

for img_path in tqdm(image_paths):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (128,128))

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)

    edges = cv2.Canny(blur, 50, 120)
    crack_pixels = np.sum(edges > 0)

    label = 1 if crack_pixels > 500 else 0
    y.append(label)
    X.append(img)

X = np.array(X)/255.0
y = np.array(y)


# ===========================================================
# 3. SPLIT DATA
# ===========================================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)


# ===========================================================
# 4. HYBRID AIS + PSO OPTIMIZER
# ===========================================================
def build_model(hp):
    model = Sequential([
        Conv2D(hp["filters"], (3,3), activation='relu', input_shape=(128,128,3)),
        MaxPooling2D(2,2),

        Conv2D(hp["filters"]*2, (3,3), activation='relu'),
        MaxPooling2D(2,2),

        Flatten(),
        Dense(hp["dense_units"], activation='relu'),
        Dropout(hp["dropout"]),
        Dense(1, activation='sigmoid')
    ])
    model.compile(
        optimizer=tf.keras.optimizers.Adam(hp["lr"]),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model


def evaluate_hp(hp):
    model = build_model(hp)
    history = model.fit(
        X_train, y_train, epochs=2, batch_size=8,
        validation_split=0.2, verbose=0
    )
    return history.history["val_accuracy"][-1]


# Search space
def random_hp():
    return {
        "lr": random.uniform(0.0001, 0.005),
        "filters": random.choice([16, 32, 48]),
        "dropout": random.uniform(0.2, 0.5),
        "dense_units": random.choice([64, 128, 256])
    }


population = [random_hp() for _ in range(6)]

best_hp = None
best_score = -1

print("[INFO] Starting AIS + PSO Optimization...")

for gen in range(5):

    print(f"\n[GENERATION {gen+1}]")

    # Evaluate population
    scores = []
    for hp in population:
        score = evaluate_hp(hp)
        scores.append(score)
        if score > best_score:
            best_score = score
            best_hp = hp

    # AIS – clone & mutate best 2
    top_idx = np.argsort(scores)[-2:]
    clones = []
    for idx in top_idx:
        clone = population[idx].copy()
        clone["lr"] *= random.uniform(0.8,1.2)
        clone["dropout"] *= random.uniform(0.9,1.1)
        clones.append(clone)

    # PSO – move others towards best
    new_population = []
    for hp in population:
        new_hp = hp.copy()
        for key in hp:
            new_hp[key] = hp[key] + random.uniform(-0.1,0.1)*(best_hp[key]-hp[key])
        new_population.append(new_hp)

    population = new_population + clones

print("\n[INFO] Best Hyperparameters Found:")
print(best_hp)


# ===========================================================
# 5. TRAIN FINAL MODEL WITH BEST HP
# ===========================================================
model = build_model(best_hp)

checkpoint = ModelCheckpoint(
    os.path.join(output_dir, "hybrid_metrosense_tzika_model.h5"),
    save_best_only=True, monitor="val_accuracy"
)

history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=8,
    validation_split=0.2,
    callbacks=[checkpoint]
)


# ===========================================================
# 6. SAVE SCALER + YAML
# ===========================================================
scaler = StandardScaler()
scaler.fit(X_train.reshape(len(X_train), -1))

with open(os.path.join(output_dir,"hybrid_metrosense_tzika_scaler.pkl"),"wb") as f:
    pickle.dump(scaler,f)

with open(os.path.join(output_dir,"hybrid_metrosense_tzika_config.yaml"),"w") as f:
    yaml.dump(best_hp,f)


# ===========================================================
# 7. PREDICTION + SAVE JSON/CSV
# ===========================================================
pred = (model.predict(X_test) > 0.5).astype("int32")

pred_json = {
    "predictions": pred.flatten().tolist(),
    "actual": y_test.tolist()
}

with open(os.path.join(output_dir,"hybrid_metrosense_tzika_prediction.json"),"w") as f:
    json.dump(pred_json,f,indent=4)

df = pd.DataFrame({
    "Actual": y_test,
    "Predicted": pred.flatten()
})
df.to_csv(os.path.join(output_dir,"hybrid_metrosense_tzika_result.csv"), index=False)


# ===========================================================
# 8. SHOW + SAVE ALL GRAPHS
# ===========================================================

# ACCURACY GRAPH
plt.figure(figsize=(6,4))
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Hybrid AIS + PSO Accuracy")
plt.show()
plt.savefig(os.path.join(output_dir,"hybrid_accuracy.png"))
plt.close()


# LOSS GRAPH
plt.figure(figsize=(6,4))
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Hybrid AIS + PSO Loss")
plt.show()
plt.savefig(os.path.join(output_dir,"hybrid_loss.png"))
plt.close()


# PREDICTION GRAPH
plt.figure(figsize=(6,4))
plt.plot(pred[:50], label="Predicted")
plt.plot(y_test[:50], label="Actual")
plt.legend()
plt.title("Hybrid Prediction Graph")
plt.show()
plt.savefig(os.path.join(output_dir,"hybrid_prediction.png"))
plt.close()


# RESULT GRAPH
plt.figure(figsize=(6,4))
df["Actual"].value_counts().plot(kind='bar')
plt.title("Hybrid Result Graph")
plt.show()
plt.savefig(os.path.join(output_dir,"hybrid_result.png"))
plt.close()


# COMPARISON GRAPH
plt.figure(figsize=(6,4))
plt.scatter(range(len(y_test)), y_test, label="Actual")
plt.scatter(range(len(pred)), pred, label="Predicted")
plt.legend()
plt.title("Hybrid Comparison Graph")
plt.show()
plt.savefig(os.path.join(output_dir,"hybrid_comparison.png"))
plt.close()


# CONFUSION MATRIX
cm = confusion_matrix(y_test, pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="coolwarm")
plt.title("Hybrid Confusion Matrix")
plt.show()
plt.savefig(os.path.join(output_dir,"hybrid_confusion_matrix.png"))
plt.close()


print("\n===================================")
print(" ALL HYBRID AIS + PSO FILES SAVED 🎉")
print("===================================\n")
```
