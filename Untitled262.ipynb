```python
import os
import cv2
import json
import yaml
import pickle
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from pdf2image import convert_from_path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint


# ===============================================================
# FIXED PATHS
# ===============================================================
pdf_path = r"C:\Users\NXTWAVE\Downloads\Underground Metro Tunnel Crack Detection\7261049\16. Tzika et al..pdf"
output_dir = r"C:\Users\NXTWAVE\Downloads\Underground Metro Tunnel Crack Detection"
images_dir = os.path.join(output_dir, "extracted_frames")
os.makedirs(images_dir, exist_ok=True)

POPPLER_PATH = r"C:\Users\NXTWAVE\Downloads\Release-25.11.0-0\poppler-25.11.0\Library\bin"


# ===============================================================
# 1. EXTRACT IMAGES FROM PDF
# ===============================================================
pages = convert_from_path(pdf_path, dpi=300, poppler_path=POPPLER_PATH)

image_paths = []
for idx, page in enumerate(pages):
    img_path = os.path.join(images_dir, f"page_{idx}.jpg")
    page.save(img_path, "JPEG")
    image_paths.append(img_path)


# ===============================================================
# 2. PREPROCESS IMAGES
# ===============================================================
X, y = [], []

for img_path in tqdm(image_paths):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (128,128))

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    edges = cv2.Canny(blur, 50, 120)
    crack_pixels = np.sum(edges > 0)

    label = 1 if crack_pixels > 500 else 0
    X.append(img)
    y.append(label)

X = np.array(X) / 255.0
y = np.array(y)


# ===============================================================
# 3. TRAIN TEST SPLIT
# ===============================================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)


# ===============================================================
# SAFE HP CLAMPING
# ===============================================================
def clamp_hp(hp):
    """Make sure all hyperparameters stay valid."""
    hp["lr"] = max(0.0001, min(0.005, hp["lr"]))
    hp["filters"] = int(max(16, min(64, hp["filters"])))
    hp["dense_units"] = int(max(32, min(512, hp["dense_units"])))
    hp["dropout"] = float(max(0.1, min(0.5, hp["dropout"])))
    return hp


# ===============================================================
# MODEL BUILDER
# ===============================================================
def build_model(hp):
    hp = clamp_hp(hp)

    model = Sequential([
        Conv2D(hp["filters"], (3,3), activation='relu', input_shape=(128,128,3)),
        MaxPooling2D(2,2),

        Conv2D(hp["filters"]*2, (3,3), activation='relu'),
        MaxPooling2D(2,2),

        Flatten(),
        Dense(hp["dense_units"], activation='relu'),
        Dropout(hp["dropout"]),
        Dense(1, activation='sigmoid')
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(hp["lr"]),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model


def evaluate_hp(hp):
    hp = clamp_hp(hp)
    model = build_model(hp)
    history = model.fit(
        X_train, y_train,
        epochs=2,
        batch_size=8,
        validation_split=0.2,
        verbose=0
    )
    return history.history["val_accuracy"][-1]


def random_hp():
    return {
        "lr": random.uniform(0.0001, 0.005),
        "filters": random.choice([16, 32, 48, 64]),
        "dropout": random.uniform(0.2, 0.5),
        "dense_units": random.choice([64, 128, 256, 512])
    }


# ===============================================================
# 4. HYBRID AIS + GWO ALGORITHM
# ===============================================================
population = [random_hp() for _ in range(6)]
best_hp = None
best_score = -1

print("[INFO] Running AIS + GWO optimization...\n")

for gen in range(5):
    print(f"\n[GENERATION {gen+1}]")

    scores = []
    for hp in population:
        score = evaluate_hp(hp)
        scores.append(score)

        if score > best_score:
            best_score = score
            best_hp = hp.copy()

    # ------------------------------
    # AIS CLONING
    # ------------------------------
    top_idx = np.argsort(scores)[-2:]
    clones = []

    for idx in top_idx:
        clone = population[idx].copy()
        clone["dropout"] *= random.uniform(0.9, 1.1)
        clone["lr"] *= random.uniform(0.8, 1.2)
        clone = clamp_hp(clone)
        clones.append(clone)

    # ------------------------------
    # GWO UPDATE
    # ------------------------------
    alpha = population[top_idx[-1]].copy()
    beta = population[top_idx[-2]].copy()
    delta = population[top_idx[0]].copy()

    new_population = []
    for hp in population:
        new_hp = hp.copy()
        for key in hp:
            A = random.uniform(-2, 2)
            C = random.uniform(0, 2)

            X1 = alpha[key] - A * abs(C * alpha[key] - hp[key])
            X2 = beta[key] - A * abs(C * beta[key] - hp[key])
            X3 = delta[key] - A * abs(C * delta[key] - hp[key])

            new_hp[key] = (X1 + X2 + X3) / 3

        new_hp = clamp_hp(new_hp)
        new_population.append(new_hp)

    population = new_population + clones


best_hp = clamp_hp(best_hp)
print("\n[INFO] Best HP:", best_hp)


# ===============================================================
# 5. TRAIN FINAL MODEL
# ===============================================================
model = build_model(best_hp)

checkpoint = ModelCheckpoint(
    os.path.join(output_dir, "ais_gwo_metrosense_tzika_model.h5"),
    monitor="val_accuracy",
    save_best_only=True
)

history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=8,
    validation_split=0.2,
    callbacks=[checkpoint]
)


# ===============================================================
# 6. SAVE SCALER + CONFIG
# ===============================================================
scaler = StandardScaler()
scaler.fit(X_train.reshape(len(X_train), -1))

with open(os.path.join(output_dir,"ais_gwo_metrosense_tzika_scaler.pkl"),"wb") as f:
    pickle.dump(scaler, f)

with open(os.path.join(output_dir,"ais_gwo_metrosense_tzika_config.yaml"),"w") as f:
    yaml.dump(best_hp, f)


# ===============================================================
# 7. PREDICTIONS + SAVE JSON + CSV
# ===============================================================
pred = (model.predict(X_test) > 0.5).astype("int32").flatten()

prediction_json = {
    "actual": y_test.tolist(),
    "predicted": pred.tolist()
}

with open(os.path.join(output_dir,"ais_gwo_metrosense_tzika_prediction.json"),"w") as f:
    json.dump(prediction_json, f, indent=4)

df = pd.DataFrame({
    "Actual": y_test,
    "Predicted": pred
})

df.to_csv(os.path.join(output_dir,"ais_gwo_metrosense_tzika_result.csv"), index=False)


# ===============================================================
# 8. PLOTS & SAVE
# ===============================================================

# Accuracy
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("AIS+GWO Accuracy")
plt.grid(True)
plt.show()
plt.savefig(os.path.join(output_dir,"ais_gwo_accuracy.png"))
plt.close()

# Loss
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("AIS+GWO Loss")
plt.grid(True)
plt.show()
plt.savefig(os.path.join(output_dir,"ais_gwo_loss.png"))
plt.close()

# Prediction Graph
plt.plot(pred[:50], label="Predicted")
plt.plot(y_test[:50], label="Actual")
plt.legend()
plt.title("AIS+GWO Prediction Graph (first 50)")
plt.grid(True)
plt.show()
plt.savefig(os.path.join(output_dir,"ais_gwo_prediction.png"))
plt.close()

# Result Graph
df["Actual"].value_counts().plot(kind='bar', alpha=0.7)
plt.title("AIS+GWO Result Graph")
plt.grid(True)
plt.show()
plt.savefig(os.path.join(output_dir,"ais_gwo_result.png"))
plt.close()

# Comparison Graph
plt.scatter(range(len(y_test)), y_test, label="Actual")
plt.scatter(range(len(pred)), pred, label="Predicted")
plt.legend()
plt.title("AIS+GWO Comparison Graph")
plt.grid(True)
plt.show()
plt.savefig(os.path.join(output_dir,"ais_gwo_comparison.png"))
plt.close()

# Confusion Matrix
cm = confusion_matrix(y_test, pred)
sns.heatmap(cm, annot=True, fmt='d', cmap="coolwarm")
plt.title("AIS+GWO Confusion Matrix")
plt.show()
plt.savefig(os.path.join(output_dir,"ais_gwo_confusion_matrix.png"))
plt.close()

print("\n==========================================")
print(" ALL AIS + GWO FILES GENERATED SUCCESSFULLY 🎉")
print("==========================================\n")
```
