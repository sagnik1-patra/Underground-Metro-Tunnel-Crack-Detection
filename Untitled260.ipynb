```python
import os
import cv2
import json
import yaml
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from pdf2image import convert_from_path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint

# ===========================================================
# PATH SETTINGS
# ===========================================================
pdf_path = r"C:\Users\NXTWAVE\Downloads\Underground Metro Tunnel Crack Detection\7261049\16. Tzika et al..pdf"
output_dir = r"C:\Users\NXTWAVE\Downloads\Underground Metro Tunnel Crack Detection"

images_dir = os.path.join(output_dir, "extracted_frames")
os.makedirs(images_dir, exist_ok=True)

POPPLER_PATH = r"C:\Users\NXTWAVE\Downloads\Release-25.11.0-0\poppler-25.11.0\Library\bin"

# ===========================================================
# 1. EXTRACT IMAGES FROM PDF
# ===========================================================
print("[INFO] Extracting images from PDF...")

pages = convert_from_path(
    pdf_path, 
    dpi=300, 
    poppler_path=POPPLER_PATH
)

image_paths = []

for idx, page in enumerate(pages):
    img_path = os.path.join(images_dir, f"page_{idx}.jpg")
    page.save(img_path, "JPEG")
    image_paths.append(img_path)

print(f"[INFO] Extracted {len(image_paths)} images successfully.")

# ===========================================================
# 2. PREPROCESS IMAGES
# ===========================================================
X = []
y = []  # Crack = 1, No Crack = 0

print("[INFO] Preprocessing images...")

for img_path in tqdm(image_paths):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (128, 128))

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)

    edges = cv2.Canny(blur, 50, 120)
    crack_pixels = np.sum(edges > 0)

    label = 1 if crack_pixels > 500 else 0
    y.append(label)
    X.append(img)

X = np.array(X)/255.0
y = np.array(y)

print("[INFO] Data shape:", X.shape)

# ===========================================================
# 3. TRAIN / TEST SPLIT
# ===========================================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# ===========================================================
# 4. CNN MODEL
# ===========================================================
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

checkpoint_path = os.path.join(output_dir, "metrosense_tzika_model.h5")
checkpoint = ModelCheckpoint(checkpoint_path, monitor='val_accuracy', save_best_only=True)

history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=8,
    validation_split=0.2,
    callbacks=[checkpoint]
)

print("[INFO] Training Completed")

# ===========================================================
# 5. SAVE SCALER + YAML CONFIG
# ===========================================================
scaler = StandardScaler()
scaler.fit(X_train.reshape(len(X_train), -1))

scaler_path = os.path.join(output_dir, "metrosense_tzika_scaler.pkl")
with open(scaler_path, "wb") as f:
    pickle.dump(scaler, f)

config = {
    "input_shape": [128,128,3],
    "optimizer": "adam",
    "loss": "binary_crossentropy",
    "dataset": "Tzika PDF Image Extract",
    "model": "Simple CNN"
}

yaml_path = os.path.join(output_dir, "metrosense_tzika_config.yaml")
with open(yaml_path, "w") as f:
    yaml.dump(config, f)

# ===========================================================
# 6. PREDICTIONS
# ===========================================================
pred = (model.predict(X_test) > 0.5).astype("int32")

# ===========================================================
# 7. SAVE PREDICTION JSON
# ===========================================================
prediction_json_path = os.path.join(output_dir, "metrosense_tzika_prediction.json")

prediction_json = {
    "total_test_samples": len(y_test),
    "predictions": pred.flatten().tolist(),
    "ground_truth": y_test.tolist()
}

with open(prediction_json_path, "w") as f:
    json.dump(prediction_json, f, indent=4)

# ===========================================================
# 8. SAVE RESULT CSV
# ===========================================================
df = pd.DataFrame({
    "Actual": y_test,
    "Predicted": pred.flatten()
})

result_csv_path = os.path.join(output_dir, "metrosense_tzika_result.csv")
df.to_csv(result_csv_path, index=False)

# ===========================================================
# 9. SHOW & SAVE ALL GRAPHS
# ===========================================================

# ACCURACY GRAPH
plt.figure(figsize=(6,4))
plt.plot(history.history["accuracy"], label="Train Accuracy")
plt.plot(history.history["val_accuracy"], label="Val Accuracy")
plt.title("Accuracy Graph")
plt.legend()
plt.grid(True)
plt.show()
plt.savefig(os.path.join(output_dir, "accuracy_graph.png"))
plt.close()

# LOSS GRAPH
plt.figure(figsize=(6,4))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.title("Loss Graph")
plt.legend()
plt.grid(True)
plt.show()
plt.savefig(os.path.join(output_dir, "loss_graph.png"))
plt.close()

# PREDICTION GRAPH
plt.figure(figsize=(6,4))
plt.plot(pred[:50], label="Predicted")
plt.plot(y_test[:50], label="Actual")
plt.title("Prediction Graph (First 50 Samples)")
plt.legend()
plt.grid(True)
plt.show()
plt.savefig(os.path.join(output_dir, "prediction_graph.png"))
plt.close()

# RESULT GRAPH (Bar chart)
plt.figure(figsize=(6,4))
df["Actual"].value_counts().plot(kind='bar', alpha=0.7, label="Actual")
plt.title("Result Graph (Actual Label Distribution)")
plt.legend()
plt.grid(True)
plt.show()
plt.savefig(os.path.join(output_dir, "result_graph.png"))
plt.close()

# COMPARISON GRAPH
plt.figure(figsize=(6,4))
plt.scatter(range(len(y_test)), y_test, label="Actual", s=20)
plt.scatter(range(len(pred)), pred, label="Predicted", s=20)
plt.title("Actual vs Predicted Comparison Graph")
plt.legend()
plt.grid(True)
plt.show()
plt.savefig(os.path.join(output_dir, "comparison_graph.png"))
plt.close()

# CONFUSION MATRIX
cm = confusion_matrix(y_test, pred)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap="coolwarm")
plt.title("Confusion Matrix")
plt.show()
plt.savefig(os.path.join(output_dir, "confusion_matrix.png"))
plt.close()

# ===========================================================
# DONE
# ===========================================================
print("\n===================================")
print(" ALL FILES SAVED & GRAPHS SHOWN 🎉")
print("===================================\n")
```
